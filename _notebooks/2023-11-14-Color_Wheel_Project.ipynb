```python
import colorsys

def rgb_to_binary(rgb):
    return f"{rgb[0]:08b} {rgb[1]:08b} {rgb[2]:08b}"

def hsv_to_rgb(h, s, v):
    return tuple(int(c * 255) for c in colorsys.hsv_to_rgb(h, s, v))

def color_wheel(angle, radius):
    angle = (angle + 360) % 360  # Ensure angle is between 0 and 360 degrees
    h = angle / 360.0
    s = min(1.0, max(0.0, radius))
    v = 1.0
    return hsv_to_rgb(h, s, v)

if __name__ == "__main__":
    print("Welcome to the Command-Line Color Wheel!")
    print("Drag the color wheel by entering the angle (0-360) and radius (0-1). Enter 'q' to quit.")

    while True:
        try:
            angle = float(input("Enter angle (0-360): "))
            if not 0 <= angle <= 360:
                raise ValueError("Angle must be between 0 and 360 degrees.")
            radius = float(input("Enter radius (0-1): "))
            if not 0 <= radius <= 1:
                raise ValueError("Radius must be between 0 and 1.")
            
            color = color_wheel(angle, radius)
            print(f"RGB: {color[0]}, {color[1]}, {color[2]}")
            print(f"Binary: {rgb_to_binary(color)}")
        
        except ValueError as ve:
            print(f"Invalid input: {ve}")
        except KeyboardInterrupt:
            print("\nExiting...")
            break
```
